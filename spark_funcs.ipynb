```python
from pyspark.sql import SparkSession
# Create a Spark session
spark = SparkSession.builder.appName("RDDExample").getOrCreate()

data_abhishek = list(range(1, 1001))
rdd_abhishek = spark.sparkContext.parallelize(data_abhishek)

a = rdd_abhishek.collect()
print("data display ->" , a)


# PySpark actions on the Abhishek RDD
print("Take:", rdd_abhishek.take(5))
print("Collect:", rdd_abhishek.collect())
print("Count:", rdd_abhishek.count())
print("First:", rdd_abhishek.first())
print("Top:", rdd_abhishek.top(5))
print("Min:", rdd_abhishek.min())
print("Max:", rdd_abhishek.max())
print("Reduce (sum):", rdd_abhishek.reduce(lambda x, y: x + y))
print("Fold (product):", rdd_abhishek.fold(1, lambda x, y: x * y))
#rdd_abhishek.saveAsTextFile("D:\\python_ai codes\\nltk_assignment_sent_analysis\\abhishek.txt")
print("Number of partitions:", rdd_abhishek.getNumPartitions())
```
