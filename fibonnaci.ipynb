**Fibonacci series with and without recursion**


```python
import time
from matplotlib import pyplot as plt
```

**creating  2 lists and then taking user input for series upto which numbers are needed**
**using time package to calculate the loop start and end time for every iteration**
**the time gives almost linear mapping in this loop with milliseconds ^3 difference**



```python
x = 0
list1 =[]
list2 =[]

def fib(x):
    input_val = int(input("please enter your number"))
    a =0
    b =1
    c =0
    start = time.time()
    while x < input_val:

        print(a)
        list2.append(x)
        c = a+b
        a = b
        b = c
        x += 1
        stop = time.time()
        duration = stop-start
        list1.append(duration)
        print(duration)


fib(x)
```

**using matplotlib to print the series in corrrespondance to time of iteration**

```python
plt.plot(list2, list1)
plt.xlabel("FIB SERIES")
plt.ylabel("TIME")
plt.grid(color='black', linestyle='-.', linewidth=0.5)
plt.show()
```

**DECLARING n-1 & n-2 recursive function to implement the series**
OBSERVATION:
since every time the function is called , for 0 ,1,1 the series takes same time with no difference as it doesnt call any recurive func. after every iteration in the loop for n-2 & n-1 for a function call the time increases gradually and then mounts to the half U shape as the number of calls being made to the recursive function is something of 2^n, or exponential complexity.With a complexity of 2^n, and an n of 40, that would be 240, or 1,099,511,627,776, or about 1 trillion operations, for first 30 numbers of series 2^30 is 1,073,741,824, which would take only about 1 second.

```python
def recur_fibo(n):
    if n <= 1:
        return n
    else:
        return(recur_fibo(n-1) + recur_fibo(n-2))

nterms = int(input("enter the number"))

# check if the number of terms is valid
if nterms <= 0:
   print("Plese enter a positive integer")
else:
   #print("Fibonacci sequence:")
   list1 =[]
   list2 =[]
   for i in range(nterms):
       start = time.time()
       list2.append(i)
       print("Fibonacci sequence:", recur_fibo(i))
       stop = time.time()
       duration = stop-start
       list1.append(duration)
       print(duration)

print(f'list 1 is {list1}')
print(f'list 2 is {list2}')
```

```python
plt.plot(list2, list1)
plt.xlabel("fib_series")
plt.ylabel("time")
plt.grid(color='black', linestyle='-.', linewidth=0.5)
plt.show()
```
